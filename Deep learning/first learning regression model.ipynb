```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

##  Generating random data

```python
observations = 1000
```

###  Creating the variables

```python
xs = np.random.uniform(-10,10,(observations,1))
zs = np.random.uniform(-10,10,(observations,1))
```

##  Transforming the variables into a matrice

```python
inputs = np.column_stack((xs,zs))
```

```python
inputs.shape
```

##  Creating the target 


```python
noise = np.random.uniform(-1,1,(observations,1))

targets = 12*xs + 5*zs - 24 + noise
targets.shape
```

##  Initializing variables

```python
init_range = 0.1
weights = np.random.uniform(-init_range,init_range,(2,1))
biases = np.random.uniform(-init_range,init_range,1)
```

##  Setting the learning rate

```python
learning_rate = 0.02
```

##  Training the model

```python
for i in range(4000):
    outputs = np.dot(inputs,weights) + biases # Calculating the outputs
    deltas = outputs - targets
    loss = np.sum(deltas**2)/2/observations # Calculating the losses
    print(f'Loss : {loss}') # Printing the loss
    deltas_scaled = deltas /observations
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled) # Recalculating the weights and the biases
    biases = biases -learning_rate * np.sum(deltas_scaled)
```

##  Printing the weights and biases

```python
print(weights,biases)
```

##  Plotting output against target for 45 degrees line

```python
plt.plot(outputs,targets)
plt.xlabel('Outputs')
plt.ylabel('Targets')
```
