##  Checking for multicolinearity

```python
data_cleaned.columns.values
```

##  Sklearn does not have to dedicated method to check this hence we use statsmodels

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables =  data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
```

```python
vif
```

###  vif should be between 1 and 5 so anything way above that is unacceptable hence we remove year

```python
data_no_multicolinearity = data_cleaned.drop(['Year'],axis =1)
```

```python
data_no_multicolinearity
```

#  Creating dummies for categorical data

```python
data_dummies = pd.get_dummies(data_no_multicolinearity, drop_first =True)
data_dummies
```

##  Rearranging the columns

```python
data_dummies.columns.values
```

```python
cols = ['Price', 'Mileage', 'EngineV', 'Brand_BMW', 'Brand_Mercedes-Benz',
       'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota',
       'Brand_Volkswagen', 'Body_hatch', 'Body_other', 'Body_sedan',
       'Body_vagon', 'Body_van', 'Engine Type_Gas', 'Engine Type_Other',
       'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_dummies[cols]
```

#  Creating the regression model itself

#  Defining inputs(x) and Target(y)

```python
inputs =data_preprocessed.drop(['Price'],axis =1)
target = data_preprocessed['Price']
```

##  Standardising

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

scaler.fit(inputs)
```

```python
scaled_inputs = scaler.transform(inputs)
```

##  Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test, = train_test_split(inputs,target, test_size=0.2,  random_state=42) 
```

#  The model

```python
reg = LinearRegression()

reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
y_hat = reg.predict(x_train)

plt.scatter(y_train, y_hat)
plt.xlabel('Training values')
plt.ylabel('Predicted values')
plt.xlim= 12
plt.ylim =12
```

##  Residual plot

###  Residual is the difference between the target and the predictions


```python
sns.distplot(y_train - y_hat)
```

####  Since the mean is zero, we are good to go even though it has quite a long tail

```python
reg.score(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg_summary = pd.DataFrame(data = inputs.columns.values, columns = ['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
data_cleaned['Brand'].unique()
```

##  Testing

```python
y_hat_test = reg.predict(x_test)

plt.scatter(y_test, y_hat_test ,alpha =0.3)
plt.xlabel('Testing values')
plt.ylabel('Predicted values')
plt.show()
```

```python
df_pf = pd.DataFrame(data = y_hat_test , columns = ['Predictions'])
df_pf
```

##  Use np.exp to change from the log to normal values

```python
df_pf = pd.DataFrame(data = np.exp(y_hat_test),columns = ["Predictions"] )
df_pf.head()
```

```python
df_pf['Target'] = np.exp(target)
```

```python
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] -df_pf['Predictions']
df_pf['Difference%'] =np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
df_pf.describe()
```

```python
df_pf.sort_values(by=['Difference%'])
```
