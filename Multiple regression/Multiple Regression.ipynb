# Creating dummy variables for categorical data

```python
raw_data = pd.read_csv('Fish.csv')
```

```python
data1 = raw_data.copy()
```

```python
data1
```

```python
data1["Species"] = data1["Species"].map({'Smelt':0,'Bream':1,'Pike':2,'Perch':3,'Roach':4,'Whitefish':5,'Parkki':6})
```

```python
data1
```

```python
data1.describe()
```

```python
x1 = data1[['Weight','Width','Species']]
y = data1['Height']
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```
